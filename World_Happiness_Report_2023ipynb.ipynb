```python
import pandas as pd
import numpy as np

df = pd.read_csv('world_happiness_2023.csv')
df.columns = ['Country','Region','Happiness_Score','GDP','Social_Support',
              'Life_Expectancy','Freedom','Generosity','Corruption']


print(f"Dataset: {len(df)} countries, {len(df.columns)} columns")
print(df.head())
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Explore the dataset before you start
print("Regions in dataset:")
print(df['Region'].value_counts())
print("\nScore range:", df['Happiness_Score'].min(), "–", df['Happiness_Score'].max())
print("\nBottom 10 countries:")
print(df.nsmallest(10, 'Happiness_Score')[['Country','Region','Happiness_Score']])
```

Task 1

```python
region_avg = (df.groupby('Region')['Happiness_Score']
              .mean()
              .reset_index()
              .sort_values('Happiness_Score'))
print(region_avg)
fig1 = px.bar(
    region_avg,
    x='Happiness_Score',
    y='Region',
    orientation='h',
    title='Western Europe Leads Global Happiness, While South Asia and Africa Lag',
    labels={'Happiness_Score': 'Average Happiness Score', 'Region': ''},
    color='Happiness_Score',
    color_continuous_scale='Blues'
)


fig1.update_layout(
    xaxis_range=[0, 10],
    plot_bgcolor='white',
    coloraxis_showscale=False,
    title_font_size=16
)
fig1.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey')
fig1.update_yaxes(showgrid=False)

fig1.show()
```

Task 2

```python

top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'
bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'
combined = pd.concat([bottom8, top8]).sort_values('Happiness_Score')
global_avg = df['Happiness_Score'].mean()
print(f"Global average: {global_avg:.2f}")


fig2 = px.bar(
    combined,
    x='Happiness_Score',
    y='Country',
    orientation='h',
    color='Group',
    color_discrete_map={'Top 8': '#2ca02c', 'Bottom 8': '#d62728'},
    title='The Happiness Gap: Top Countries Score More Than Triple the Bottom Countries',
    labels={'Happiness_Score': 'Happiness Score', 'Country': ''}
)


fig2.update_layout(
    xaxis_range=[0, 10],
    plot_bgcolor='white',
    showlegend=False,
    title_font_size=16
)


fig2.add_vline(
    x=global_avg,
    line_width=2,
    line_dash="dash",
    line_color="grey",
    annotation_text=f"Global Average ({global_avg:.2f})",
    annotation_position="top right"
)


fig2.add_hline(y=7.5, line_width=1, line_color="black")

fig2.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey')
fig2.update_yaxes(showgrid=False)

fig2.show()
```

Task 3

```python

target_regions = 'Western Europe|Latin America|East Asia|Sub-Saharan Africa|South Asia'
stretch_df = df[df['Region'].str.contains(target_regions, na=False, case=False)]


stretch_avg = stretch_df.groupby('Region')[['GDP', 'Freedom']].mean().reset_index()


stretch_melted = stretch_avg.melt(
    id_vars='Region',
    value_vars=['GDP', 'Freedom'],
    var_name='Factor',
    value_name='Average_Score'
)

fig3 = px.bar(
    stretch_melted,
    x='Region',
    y='Average_Score',
    color='Factor',
    barmode='group',
    title='Economic Wealth (GDP) Varies More Drastically By Region Than Perceived Freedom',
    labels={'Average_Score': 'Average Score', 'Region': ''},
    color_discrete_map={'GDP': '#1f77b4', 'Freedom': '#ff7f0e'}
)


fig3.update_layout(
    plot_bgcolor='white',
    yaxis_range=[0, stretch_melted['Average_Score'].max() * 1.1]
)
fig3.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey')
fig3.update_xaxes(showgrid=False)

fig3.show()
```
